```python
y = df['median_house_value']
y.head()
```

```python
x = df.drop('median_house_value', axis=1)
x.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

x_train_simple = x_train[['median_income']]
x_test_simple = x_test[['median_income']]

model_simple = LinearRegression()
model_simple.fit(x_train_simple, y_train)
```

```python

y_pred_simple = model_simple.predict(x_test_simple)

print("Simple Linear Regression:")
print("R² Score:", r2_score(y_test, y_pred_simple))
print("MSE:", mean_squared_error(y_test, y_pred_simple))
rmse_simple = np.sqrt(mse_simple)
print("RMSE:", rmse_simple)
```

```python
model_multi = LinearRegression()
model_multi.fit(x_train, y_train)
```

```python
y_pred_multi = model_multi.predict(x_test)

print("Multiple Linear Regression:")
print("R² Score:", r2_score(y_test, y_pred_multi))
print("MSE:", mean_squared_error(y_test, y_pred_multi))
rmse_multi = np.sqrt(mse_multi)
print("RMSE:", rmse_multi)
```

```python
import pandas as pd
from sklearn.metrics import r2_score, mean_squared_error

r2_simple = r2_score(y_test, y_pred_simple)
mse_simple = mean_squared_error(y_test, y_pred_simple)

r2_multi = r2_score(y_test, y_pred_multi)
mse_multi = mean_squared_error(y_test, y_pred_multi)

comparison_df = pd.DataFrame({
    'Model': ['Simple Linear Regression', 'Multiple Linear Regression'],
    'R² Score': [r2_simple, r2_multi],
    'MSE': [mse_simple, mse_multi],
    'RMSE': [rmse_simple, rmse_multi]
})
```

```python
from tabulate import tabulate

print(tabulate(comparison_df, headers='keys', tablefmt='pretty', showindex=False))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred_multi, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red')  # perfect prediction line
plt.xlabel("Actual Median House Value")
plt.ylabel("Predicted Median House Value")
plt.title("Actual vs Predicted (Multiple Linear Regression)")
plt.show()
```

```python
residuals = y_test - y_pred_multi
sns.histplot(residuals, kde=True)
plt.title("Residuals Distribution (Multiple Linear Regression)")
plt.xlabel("Residuals")
plt.show()
```

**Objective**:
Predict the median house value in California districts using regression models.

**Dataset Overview**:
Includes features like `median_income`, `total_rooms`, `households`, `latitude`, and `ocean_proximity`.

* Strong correlation between `median_income` and house value.
